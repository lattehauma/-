```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
tips=sns.load_dataset('tips')
plt.subplots()
```

```python
df1=pd.DataFrame(
    {'age':[30,23,26],
    'sex':['M','W','M'],
    'height':[178,162,172],
    'weight':[62,53,61]
     }
)
print(df1)
df1=pd.melt(df1, id_vars=['age','sex'], var_name='group', value_name='data')
print(df1)
```

subplot그리기 (한 번에 여러 개의 그래프 그리기)

문자열의 변환과 그래프 작성하기

```python
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
tips['sex_color']=tips['sex'].apply(recode_sex)
ax=plt.scatter(
    x=tips['total_bill'], y=tips['tip'],
    s=tips['size']*10, c=tips['sex_color'], alpha=0.5 #알파는 투명도
)
```
